```python
import pandas as pd
```

```python
datos_cebollitas = pd.read_csv("./data/partidos_cebollitas.csv")
datos_cebollitas.head()
```

```python
datos_cebollitas.info()
```

```python
datos_cebollitas.isnull().sum()
datos_cebollitas["goles_local"] = datos_cebollitas["goles_local"].fillna(datos_cebollitas["goles_local"].mean())
```

```python
#one hot encoding
datos_preparados = pd.get_dummies(datos_cebollitas, columns=["equipo_visitante", "equipo_local"])
datos_preparados.head()
```

```python
#Remove duplicates
print(datos_preparados.shape)
datos_preparados.drop_duplicates(inplace=True)
print(datos_preparados.shape)
```

```python
datos_preparados["fecha_partido"] = pd.to_datetime(datos_preparados["fecha_partido"])
datos_preparados.head()
```
